# 1. Regresión lineal: Conversión de °C a °F

Vamos a crear un modelo de regresión lineal que aprenda la relación entre la temperatura en grados Celsius (°C) y Fahrenheit (°F).

Sabemos que la fórmula para convertir °C a °F es:

$$°F = (1.8 \times °C) + 32$$

Sin embargo, simularemos que no conocemos la fórmula y queremos que el modelo la aprenda a partir de ciertos datos.

## Importar las librerías

```python
import numpy as np

```

````python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
````

## Creacion de los datos

- C represnta las entradas (temperaturas en grados C)
- F representa las salidas o etiquetas (temperaturas en grados F)
- El metodo .reshape(-1,1) transforma el arreglo de una **Matriz columna** (tamano n x 1) con una con una sola caracteristica, como espera scikit-learn.

Nota: C y F son los datos de entrenaminiento, equivalente C a la entrada y F la salida.

````python
# Datos

C = np.array([0, 10, 20,30,40,50]).reshape(-1,1)
F = np.array([32,50,68,86,104,122])
````

## Creacion y entrenamiento del modelo
- LinearRegression() crea una instancia del modelo
- .fit(C,F) entrena el modelo para encontrar los coeficientes  $\beta_0$ (interseccion) y $\beta_1$ (pendiente). Es decir, la ecuacion del modelo aprendido sera:
- $\hat{F} = \beta_0 + \beta_1 \times {}^{\circ}\mathrm{C}$

````python
# Crear el modelo y entrenar el modelo
modelo = LinearRegression()
modelo.fit(C, F)
````

## Prediccion y evaluacion del modelo

Despues de entrenar el modelo con los datos de entrenamiento (modelo.fit(C, F)), podemos usarlo para predecir valores y probar que tan bien aprendio la relacion entre grados C y F grados.
- **pred** es arreglo con las salidas estimadas; aqui se predice sobre los mismos datos de entrenaminiento para visualizar el ajuste.

````python
# Prediccion
x_test = np.array([25, 23, 27, 38, 54, 105]).reshape(-1,1)
pred = modelo.predict(x_test)
print(pred)
````

## Resultados del modelo
- **modelo.coef_**, Pendiente o coeficiente $\beta_1$
- **modelo.intercept_**, Interseccion o constante $\beta_0$


````python
## Resultados
print("Pendiente (w1)", modelo.coef_[0])
print("Interseccion (w0)", modelo.intercept_)
print(f"Ecuacion del modelo: F {modelo.coef_[0]:.2f} *C + {modelo.intercept_:.2f} ")
````

## Visualizacion

````python
# visualizacion
plt.scatter(C, F, color='blue', label='Datos reales')
plt.plot(x_test, pred, color='red', label = 'Linea de regresion')
plt.xlabel('Temperatura (grados C)')
plt.ylabel('Temperatura (grados F)')
plt.title("Regresion lineal grados C a grados F")
plt.legend()
plt.show()
````

## Evaluacion del modelo 
Una vez que el modelo ha sido entrenado, vamos a ponerlo a prueba con un nuevo valor de temperatura en grados Celcius que no formo parte del conjunto de entrenamiento.
El objetivo es comprobar que tan preciso es el modelo al predecir su equivalente en grados Fahrenheit, y calcular la diferencia entre el valor real y el valor predicho.
A partir de esta comparacion obtendremos:
- El error absoluto, que mide cuanto se desvia el modelo en grados F
- El error relativo, que indica el porcentaje de error respecto al valor real.

````python
import numpy as np

# Valor nuevo
C_nuevo = 200.0
X_nuevo = np.array([[C_nuevo]]) # skleant espera 2D

# Prediccion del modelo ya entrenado (ariable: modelo)
F_pred = modelo.predict(X_nuevo)[0]

# Valor real (temperatura verdadera)
F_real = (1.8 * C_nuevo) + 32

# errores
error_abs = abs(F_real - F_pred)
error_rel = error_abs / abs(F_real) if F_real != 0 else 0.0

print(f"C = {C_nuevo} grados C")
print(f"F_predicho = {F_pred:.6f} F")
print(f"F_real = {F_real:.6f} F")
print(f"Error absoluto = {error_abs:.6e} F")
print(f"Error relativo = {100*error_rel:.6e} %")
````

esta funcion ***divide aleatoriamente** los datos en dos grupos:
- 80% entrenamiento
- 20% prueba
Esa "aleatoriedad" se genera con un numero aleatorio de inicio, llamado semilla o random seed.
El argumento random_state **fija la semilla del generador aleatorio** que usa Scikit-Learn para hacer la division.
En otras palabras:
- Si no pones **random_state**, la division cambiara cada vez que ejecutes el codigo.
- Si no pones un numero fijo (por ejemplo 42), la division sera siempre la misma.
Asi garantizas reproducibilidad: tu, tus estudiante o cualquier persona que ejecute el mismo codigo obtendra exactamente los mismos conjuntos de entrenamiento y prueba.
