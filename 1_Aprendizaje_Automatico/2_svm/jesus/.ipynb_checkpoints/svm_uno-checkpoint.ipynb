## 1. Introducción

Este reporte presenta un análisis práctico de las Máquinas de Vectores de Soporte (Support Vector Machines - SVM) utilizando un kernel lineal. El objetivo es demostrar cómo SVM encuentra el hiperplano óptimo que separa dos clases de datos con el máximo margen posible.

### Objetivos:
- Implementar un clasificador SVM con kernel lineal
- Visualizar la frontera de decisión y los márgenes
- Identificar los vectores de soporte
- Analizar el comportamiento del modelo

## 2. Marco Teórico

### 2.1 ¿Qué es SVM?

Las Máquinas de Vectores de Soporte son algoritmos de aprendizaje supervisado utilizados para clasificación y regresión. Su objetivo principal es encontrar el hiperplano que mejor separa las clases en el espacio de características, maximizando el margen entre las clases.

### 2.2 Conceptos Clave

- **Hiperplano de decisión**: Frontera que separa las clases (línea continua en la visualización)
- **Vectores de soporte**: Puntos más cercanos al hiperplano que determinan su posición (círculos negros)
- **Margen**: Distancia entre el hiperplano y los vectores de soporte (líneas punteadas)
- **Parámetro C**: Controla el balance entre margen máximo y error de clasificación (C=1000 = margen duro)

## 3. Implementación

### 3.1 Importación de Bibliotecas

```python
# Importación de bibliotecas necesarias
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
from sklearn.inspection import DecisionBoundaryDisplay
```

### 3.2 Generación de Datos

Creamos un conjunto de datos sintético con 40 puntos separables en dos clases utilizando `make_blobs`.

```python
# Creamos 40 puntos separables en dos clases
X, y = make_blobs(n_samples=40, centers=2, random_state=6)

print("Características (X):")
print(X)
print("\nEtiquetas (y):")
print(y)
print(f"\nForma de X: {X.shape}")
print(f"Clases únicas: {set(y)}")
```

### 3.3 Entrenamiento del Modelo SVM

Configuramos y entrenamos un clasificador SVM con las siguientes características:
- **Kernel**: Lineal (para datos linealmente separables)
- **C**: 1000 (parámetro de regularización alto = margen duro, penaliza fuertemente errores)

```python
# Entrenamos el modelo SVM con kernel lineal
# C=1000 significa que no regularizamos (margen duro)
clf = svm.SVC(kernel="linear", C=1000)
clf.fit(X, y)

print(f"Número de vectores de soporte: {len(clf.support_vectors_)}")
print(f"Índices de vectores de soporte: {clf.support_}")
print(f"\nVectores de soporte por clase: {clf.n_support_}")
```

## 4. Visualización de Resultados

### 4.1 Gráfica de la Frontera de Decisión

Visualizamos:
- **Puntos de datos**: Coloreados según su clase
- **Hiperplano de decisión**: Línea sólida (nivel 0)
- **Márgenes**: Líneas punteadas (niveles -1 y +1)
- **Vectores de soporte**: Círculos negros alrededor de los puntos críticos

```python
# Creamos la visualización
plt.figure(figsize=(10, 8))

# Graficamos los puntos de datos
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# Obtenemos el eje actual
ax = plt.gca()

# Graficamos la frontera de decisión y los márgenes
DecisionBoundaryDisplay.from_estimator(
    clf,
    X,
    plot_method="contour",
    colors="k",
    levels=[-1, 0, 1],  # -1 y 1 son los márgenes, 0 es el hiperplano
    alpha=0.5,
    linestyles=["--", "-", "--"],  # punteado para márgenes, sólido para hiperplano
    ax=ax,
)

# Resaltamos los vectores de soporte
ax.scatter(
    clf.support_vectors_[:, 0],
    clf.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)

plt.title("SVM con Kernel Lineal - Frontera de Decisión y Vectores de Soporte", 
          fontsize=14, fontweight='bold')
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.grid(True, alpha=0.3)
plt.show()
```

## 5. Análisis de Resultados

### 5.1 Interpretación de la Visualización

En la gráfica podemos observar:

1. **Separación lineal**: El hiperplano (línea sólida negra) separa perfectamente ambas clases
2. **Márgenes máximos**: Las líneas punteadas muestran el margen de separación
3. **Vectores de soporte**: Los puntos marcados con círculos negros son los que definen la frontera
4. **Margen duro**: Con C=1000, el modelo busca un margen perfecto sin permitir errores

### 5.2 Características del Modelo

```python
# Información adicional del modelo
print("=" * 60)
print("RESUMEN DEL MODELO SVM")
print("=" * 60)
print(f"\nKernel utilizado: {clf.kernel}")
print(f"Parámetro C: {clf.C}")
print(f"\nNúmero total de muestras: {len(X)}")
print(f"Número de vectores de soporte: {len(clf.support_vectors_)}")
print(f"Porcentaje de vectores de soporte: {len(clf.support_vectors_)/len(X)*100:.2f}%")
print(f"\nVectores de soporte por clase:")
print(f"  Clase 0: {clf.n_support_[0]}")
print(f"  Clase 1: {clf.n_support_[1]}")
print(f"\nPrecisión en entrenamiento: {clf.score(X, y)*100:.2f}%")
print("=" * 60)
```

## 6. Conclusiones

En este ejercicio hemos implementado exitosamente un clasificador SVM con kernel lineal que demuestra los conceptos fundamentales de este algoritmo:

1. **Separación óptima**: El algoritmo encontró el hiperplano que maximiza el margen entre clases
2. **Vectores de soporte**: Solo un subconjunto de los puntos (los más cercanos a la frontera) determinan la solución
3. **Margen duro**: Con C=1000, el modelo no tolera errores de clasificación en los datos de entrenamiento
4. **Eficiencia**: SVM solo necesita almacenar los vectores de soporte, no todos los datos
