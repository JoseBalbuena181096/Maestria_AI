## 1. Importación de Librerías

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.inspection import DecisionBoundaryDisplay

plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (10, 6)
import warnings
warnings.filterwarnings('ignore')
print("✓ Librerías cargadas")
```

# EJEMPLO 1: Hiperplano de Separación
## 2. Generación de Datos
Creamos datos sintéticos linealmente separables con `make_blobs`.

```python
A, etiquetas = make_blobs(n_samples=40, centers=2, random_state=6)
print(f"Datos generados: {A.shape}")
print(f"Clase 0: {np.sum(etiquetas==0)}, Clase 1: {np.sum(etiquetas==1)}")

plt.figure(figsize=(8,6))
plt.scatter(A[:,0], A[:,1], c=etiquetas, cmap='viridis', s=80, edgecolors='black')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Datos Sintéticos - 2 Clases')
plt.show()
```

## 3.1 Entrenamiento SVM Base
SVM con kernel lineal y C=0.1 (margen amplio).

```python
modelo_base = svm.SVC(kernel='linear', C=0.1)
modelo_base.fit(A, etiquetas)

print(f"Vectores de soporte: {len(modelo_base.support_vectors_)}")
print(f"Coeficientes (w): {modelo_base.coef_}")
print(f"Intercepto (b): {modelo_base.intercept_}")
print(f"Precisión: {modelo_base.score(A, etiquetas)*100:.2f}%")
```

## 4.1 Visualización del Hiperplano

```python
plt.figure(figsize=(10,7))
plt.scatter(A[:,0], A[:,1], c=etiquetas, cmap='viridis', s=60, edgecolors='black')
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(modelo_base, A, plot_method="contour",
    colors="purple", levels=[-1,0,1], alpha=0.5, linestyles=["--","-","--"], ax=ax)
ax.scatter(modelo_base.support_vectors_[:,0], modelo_base.support_vectors_[:,1],
    s=200, facecolors="none", edgecolors="red", linewidth=2, label="Vectores de soporte")
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('SVM - Hiperplano y Márgenes')
plt.legend()
plt.show()
```

## 3.2 Entrenamiento SVM Base
SVM con kernel lineal y C = 10 (margen estrecho).

```python
modelo_base = svm.SVC(kernel='linear', C=10)
modelo_base.fit(A, etiquetas)

print(f"Vectores de soporte: {len(modelo_base.support_vectors_)}")
print(f"Coeficientes (w): {modelo_base.coef_}")
print(f"Intercepto (b): {modelo_base.intercept_}")
print(f"Precisión: {modelo_base.score(A, etiquetas)*100:.2f}%")
```

## 4.2 Visualización del Hiperplano

```python
plt.figure(figsize=(10,7))
plt.scatter(A[:,0], A[:,1], c=etiquetas, cmap='viridis', s=60, edgecolors='black')
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(modelo_base, A, plot_method="contour",
    colors="purple", levels=[-1,0,1], alpha=0.5, linestyles=["--","-","--"], ax=ax)
ax.scatter(modelo_base.support_vectors_[:,0], modelo_base.support_vectors_[:,1],
    s=200, facecolors="none", edgecolors="red", linewidth=2, label="Vectores de soporte")
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('SVM - Hiperplano y Márgenes')
plt.legend()
plt.show()
```

## 3.3 Entrenamiento SVM Base
SVM con kernel RBF y C = 0.1

```python
# Con kernel RBF, solo puedes acceder a cierta información
modelo_base = svm.SVC(kernel='rbf', C=0.1)
modelo_base.fit(A, etiquetas)

print(f"Vectores de soporte: {len(modelo_base.support_vectors_)}")
print(f"Índices de vectores de soporte: {modelo_base.support_}")
print(f"Intercepto (b): {modelo_base.intercept_}")
print(f"Precisión: {modelo_base.score(A, etiquetas)*100:.2f}%")

# También puedes ver otros atributos útiles:
print(f"Número de vectores de soporte por clase: {modelo_base.n_support_}")
print(f"Coeficientes duales: {modelo_base.dual_coef_}")
```

## 4.3 Visualización del Hiperplano

```python
plt.figure(figsize=(10,7))
plt.scatter(A[:,0], A[:,1], c=etiquetas, cmap='plasma', s=60, edgecolors='navy')
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(modelo_base, A, plot_method="contour",
    colors="green", levels=[-1,0,1], alpha=0.5, linestyles=["--","-","--"], ax=ax)
ax.scatter(modelo_base.support_vectors_[:,0], modelo_base.support_vectors_[:,1],
    s=200, facecolors="none", edgecolors="orange", linewidth=2, label="Vectores de soporte")
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('SVM - Hiperplano y Márgenes (RBF)')
plt.legend()
plt.show()
```

## Análisis del Ejemplo 1: Hiperplano de Separación (Datos Lineales)

El Ejemplo 1 utiliza datos **sintéticos linealmente separables** (20 de Clase 0 y 20 de Clase 1)
### 1.1. Efecto del Parámetro $C$ (Kernel Lineal)

El parámetro $C$ controla la penalización por errores y afecta el margen y el número de Vectores de Soporte (SV).

| Configuración | $C$ | Precisión (Train) | Vectores de Soporte (SV) | Observación |
| :---: | :---: | :---: | :---: | :---: |
| `linear_C0.1` | **0.1** | 100.00% | **6** | Margen más **amplio** (suave). |
| `linear_C10` | **10** | 100.00% | **3**  | Margen más **estrecho** (estricto). |

**Conclusión sobre $C$ (Datos Lineales):**
En escenarios donde los datos son completamente separables, el **kernel lineal** ofrece resultados óptimos. Utilizando un **valor reducido de $C$ ($0.1$)** se obtiene un margen de separación más flexible con mayor cantidad de SV. Por el contrario, un **valor elevado de $C$ ($10$)** produce un margen de separación más rígido con menor cantidad de SV.

### 1.2. Efecto del Tipo de Kernel (RBF)

Al implementar un kernel no lineal con datos que son linealmente separables, el comportamiento de la frontera cambia:

| Configuración | $C$ | Precisión (Train) | Vectores de Soporte (SV) | Forma de la Frontera |
| :---: | :---: | :---: | :---: | :---: |
| `rbf_C0.1` | 0.1 | 100.00% | **40**  | Frontera **Radial** (curvatura flexible). Cada punto actúa como SV. |
| `rbf_C10` | 10 | 100.00% | **5**  | Frontera **Radial** con ajuste más preciso. |

**Conclusión sobre Kernels (Datos Lineales):**
El **Kernel RBF** produce fronteras con forma curva mediante funciones de base radial. Si bien logra clasificar correctamente los datos, añade una complejidad innecesaria. En situaciones con datos linealmente separables, **utilizar el kernel lineal resulta más apropiado por su simplicidad y eficiencia**.

# EJEMPLO 2: Dataset Real ESL.mixture
## 5. Carga de Datos Reales

```python
url = 'https://raw.githubusercontent.com/JoaquinAmatRodrigo/Estadistica-machine-learning-python/master/data/ESL.mixture.csv'
datos = pd.read_csv(url)
print(f"Dataset: {datos.shape}")
print(datos.head())
print(f"\nDistribución: {datos['y'].value_counts()}")

caracteristicas = datos[['X1','X2']].values
clases = datos['y'].values

plt.figure(figsize=(8,6))
plt.scatter(caracteristicas[:,0], caracteristicas[:,1], c=clases, cmap='RdYlBu', alpha=0.7, s=55, edgecolors='darkgreen')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Dataset ESL.mixture')
plt.show()
```

## 6. División de Datos

```python
A_train, A_test, etiq_train, etiq_test = train_test_split(caracteristicas, clases, test_size=0.2, random_state=42, stratify=clases)
print(f"Entrenamiento: {A_train.shape[0]}, Prueba: {A_test.shape[0]}")
```

# 1.3 MODIFICACIONES: Diferentes C y Kernels
## 7. Entrenamiento con Todas las Configuraciones

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
valores_C = [0.1, 10]
resultados = {}

for C in valores_C:
    for kernel in kernels:
        clave = f"{kernel}_C{C}"
        modelo = svm.SVC(kernel=kernel, C=C, gamma='scale')
        modelo.fit(A_train, etiq_train)
        
        prec_train = modelo.score(A_train, etiq_train)
        prec_test = modelo.score(A_test, etiq_test)
        num_sv = len(modelo.support_vectors_)
        
        resultados[clave] = {
            'modelo': modelo, 'kernel': kernel, 'C': C,
            'train_acc': prec_train, 'test_acc': prec_test, 'n_sv': num_sv
        }
        print(f"{clave}: Entrenamiento={prec_train:.3f}, Prueba={prec_test:.3f}, SV={num_sv}")

print("\n✓ Entrenamiento completado")
```

## 8. Tabla Comparativa

```python
df_resultados = pd.DataFrame([{
    'Config': k, 'Kernel': v['kernel'], 'C': v['C'],
    'Entrenamiento': v['train_acc'], 'Prueba': v['test_acc'], 'SV': v['n_sv']
} for k,v in resultados.items()])
df_resultados = df_resultados.sort_values('Prueba', ascending=False)
print("\nRESULTADOS COMPLETOS:")
print(df_resultados.to_string(index=False))
```

## 9. Visualización Comparativa

```python
fig, ax = plt.subplots(figsize=(12,6))
x = np.arange(len(df_resultados))
ancho = 0.35
ax.bar(x-ancho/2, df_resultados['Entrenamiento'], ancho, label='Entrenamiento', alpha=0.8, color='forestgreen')
ax.bar(x+ancho/2, df_resultados['Prueba'], ancho, label='Prueba', alpha=0.8, color='steelblue')
ax.set_xlabel('Configuración')
ax.set_ylabel('Precisión')
ax.set_title('Comparación de Precisión por Configuración')
ax.set_xticks(x)
ax.set_xticklabels(df_resultados['Config'], rotation=45, ha='right')
ax.legend()
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

## 10. Fronteras de Decisión - Todas las Configuraciones

```python
fig, axes = plt.subplots(2, 4, figsize=(20,10))
axes = axes.ravel()

for idx, (clave, info) in enumerate(resultados.items()):
    ax = axes[idx]
    DecisionBoundaryDisplay.from_estimator(info['modelo'], A_test, alpha=0.4, ax=ax,
        response_method="predict", plot_method="pcolormesh", cmap='RdYlBu')
    ax.scatter(A_test[:,0], A_test[:,1], c=etiq_test, s=35, edgecolors='darkgreen', cmap='RdYlBu')
    ax.set_title(f"{info['kernel']}, C={info['C']}\nPrecisión={info['test_acc']:.3f}, SV={info['n_sv']}")
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')

plt.suptitle('Fronteras de Decisión - Todas las Configuraciones', fontsize=16, y=1.00)
plt.tight_layout()
plt.show()
```

# 1.4 ANÁLISIS DE CAMBIOS
## 11. Efecto del Parámetro C

```python
print("="*70)
print("ANÁLISIS: EFECTO DEL PARÁMETRO C")
print("="*70)

for kernel in kernels:
    bajo = resultados[f"{kernel}_C0.1"]
    alto = resultados[f"{kernel}_C10"]
    print(f"\n{kernel.upper()}:")
    print(f"  C=0.1: Prueba={bajo['test_acc']:.3f}, SV={bajo['n_sv']}")
    print(f"  C=10:  Prueba={alto['test_acc']:.3f}, SV={alto['n_sv']}")
    print(f"  Cambio SV: {alto['n_sv']-bajo['n_sv']:+d}")

print("""
\nOBSERVACIONES:
- C REDUCIDO (0.1): Zona de margen flexible, incremento de vectores de soporte, reducción de sobreajuste
- C ELEVADO (10): Zona de margen restrictiva, disminución de vectores de soporte, incremento del riesgo de sobreajuste
""")
```

## 12. Análisis por Kernel

📈 Análisis del Ejemplo 2: Dataset Real ESL.mixture (Datos No Lineales)

En el Ejemplo 2 trabajamos con el **Dataset ESL.mixture**, caracterizado por tener una **estructura no lineal**.

### Tabla de Resultados por Precisión en Prueba

| Config | Kernel | $C$ | Precisión (Train) | Precisión (Test) | SV |
| :---: | :---: | :---: | :---: | :---: | :---: |
| **rbf\_C10** | **RBF** | **10.0** | **0.819** | **0.850** | **84** |
| linear\_C0.1 | Linear | 0.1 | 0.719 | 0.775 | 112 |
| linear\_C10 | Linear | 10.0 | 0.713 | 0.775 | 101 |
| sigmoid\_C0.1 | Sigmoid | 0.1 | 0.713 | 0.775 | 124 |
| rbf\_C0.1 | RBF | 0.1 | 0.744 | 0.750 | 140 |
| poly\_C0.1 | Poly | 0.1 | 0.650 | 0.600 | 123 |
| poly\_C10 | Poly | 10.0 | 0.662 | 0.600 | 115 |
| sigmoid\_C10 | Sigmoid | 10.0 | 0.637 | 0.600 | 62 |

### Impacto de la Selección de Kernel (Estructura No Lineal)

* **RBF:** Demuestra ser el **más adaptable** y proporcionó los mejores resultados en datos con patrones no lineales. Alcanzó el **rendimiento superior (0.850)**, creando fronteras flexibles y complejas.
* **Linear:** Ofrece interpretabilidad y simplicidad, sin embargo su **límite de decisión rectilíneo** restringe su capacidad de precisión en estructuras no lineales (0.775).
* **Poly y Sigmoid:** Registraron el desempeño más bajo en esta configuración de datos (0.600). El Sigmoid presenta aplicaciones limitadas y menor estabilidad.

### Influencia del Parámetro $C$ (Estructura No Lineal)

* **RBF:** Incrementar **$C$ a 10** elevó la precisión a 0.850. Esto sugiere que establecer un margen más restrictivo fue beneficioso para modelar la complejidad inherente a los datos no lineales.
* **Sigmoid:** Aumentar **$C$ a 10** disminuyó considerablemente la precisión de 0.775 a 0.600, evidenciando que el margen restrictivo provocó **problemas de ajuste excesivo** o desempeño inadecuado con este kernel.
* **Observación General:** Valores elevados de $C$ disminuyen la cantidad de Vectores de Soporte en todas las configuraciones de kernel.

## 13. Matrices de Confusión (Top 4)

```python
top4 = df_resultados.head(4)['Config'].values
fig, axes = plt.subplots(2,2, figsize=(12,10))
axes = axes.ravel()

for idx, config in enumerate(top4):
    info = resultados[config]
    predicciones = info['modelo'].predict(A_test)
    matriz_conf = confusion_matrix(etiq_test, predicciones)
    sns.heatmap(matriz_conf, annot=True, fmt='d', cmap='YlGnBu', ax=axes[idx])
    axes[idx].set_title(f"{config}\nPrecisión={info['test_acc']:.3f}")
    axes[idx].set_ylabel('Real')
    axes[idx].set_xlabel('Predicción')

plt.suptitle('Matrices de Confusión - Top 4 Modelos', fontsize=14)
plt.tight_layout()
plt.show()
```

## 14. CONCLUSIONES


## Conclusiones Finales - Support Vector Machines

### Influencia de los Hiperparámetros

1.  **Impacto del Parámetro $C$:**
    * Valores reducidos de $C$ $\to$ Zona de margen **flexible**, **incremento de SV**, capacidad de generalización mejorada.
    * Valores elevados de $C$ $\to$ Zona de margen **restrictiva**, **reducción de SV**, probabilidad de sobreajuste.
2.  **Evaluación de Kernels:**
    * **RBF:** Presenta mayor versatilidad, ideal para estructuras no lineales.
    * **Linear:** Directo, computacionalmente eficiente, facilita la interpretación (apropiado para estructuras lineales).
    * **Poly:** Efectivo para relaciones polinómicas particulares.

### Configuración Óptima

La combinación que registró mejor desempeño en el dataset con estructura no lineal (`ESL.mixture`) fue **RBF con $C=10$**, alcanzando una precisión en prueba del **85.0%**.

### Sugerencias Prácticas

* Iniciar experimentación con el **Kernel RBF** como referencia inicial (*baseline*).
* Implementar **validación cruzada** para determinar el valor óptimo de $C$.
* Vigilar la discrepancia entre precisión de entrenamiento y prueba para identificar sobreajuste.
