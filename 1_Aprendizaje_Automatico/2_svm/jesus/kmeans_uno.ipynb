# Clasificación de Iris con K-Nearest Neighbors (KNN)
## Análisis Comparativo de Pesos en KNN

### 1. Introducción

Este notebook presenta un análisis de clasificación del dataset Iris utilizando el algoritmo K-Nearest Neighbors (KNN). El objetivo principal es comparar el rendimiento del clasificador utilizando dos esquemas de ponderación diferentes: **uniform** y **distance**.

El dataset Iris es un conjunto de datos clásico en machine learning que contiene mediciones de flores de tres especies diferentes de iris. Para este análisis, utilizaremos únicamente dos características: longitud y ancho del sépalo.

---
### 2. Importación de Bibliotecas

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np
```

---
### 3. Carga y Preparación de Datos

**Características seleccionadas:**
- **Sepal length (cm)**: Longitud del sépalo en centímetros
- **Sepal width (cm)**: Ancho del sépalo en centímetros

**Clases objetivo:**
- Setosa (0)
- Versicolor (1)
- Virginica (2)

```python
# Cargar el dataset Iris
iris = load_iris(as_frame=True)

# Seleccionar solo dos características para facilitar la visualización
X = iris.data[["sepal length (cm)", "sepal width (cm)"]]
y = iris.target

print(f"Dimensiones del dataset: {X.shape}")
print(f"Clases objetivo: {iris.target_names}")
print(f"\nPrimeras filas del dataset:")
print(X.head())
```

---
### 4. División de Datos

Dividimos los datos en conjuntos de entrenamiento y prueba. El parámetro `stratify=y` asegura que la proporción de clases se mantenga en ambos conjuntos.

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=0
)

print(f"Tamaño del conjunto de entrenamiento: {X_train.shape[0]}")
print(f"Tamaño del conjunto de prueba: {X_test.shape[0]}")
print(f"\nDistribución de clases en entrenamiento:")
print(y_train.value_counts().sort_index())
```

---
### 5. Construcción del Pipeline

**Componentes del pipeline:**

1. **StandardScaler**: Normaliza las características para que tengan media 0 y desviación estándar 1. Esto es importante en KNN porque el algoritmo es sensible a la escala de las características.

2. **KNeighborsClassifier**: Clasificador KNN con k=11 vecinos. Este valor de k fue seleccionado para proporcionar un balance entre sesgo y varianza.

```python
# Crear un pipeline que incluye:
# 1. Escalado de características (StandardScaler)
# 2. Clasificador KNN con k=11 vecinos
clf = Pipeline(
    steps=[
        ("scaler", StandardScaler()), 
        ("knn", KNeighborsClassifier(n_neighbors=11))
    ]
)

print("Pipeline creado exitosamente:")
print(clf)
```

---
### 6. Comparación de Métodos de Ponderación

#### 6.1 Métodos de Ponderación

- **Uniform**: Todos los vecinos tienen el mismo peso en la votación.
- **Distance**: Los vecinos más cercanos tienen mayor peso. El peso es inversamente proporcional a la distancia.

```python
# Evaluar el rendimiento con diferentes esquemas de ponderación
print("Evaluación de modelos:\n")
for weights in ("uniform", "distance"):
    clf.set_params(knn__weights=weights).fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    print(f"Accuracy con weights='{weights}': {score:.4f}")
```

---
### 7. Visualización de Fronteras de Decisión

Las fronteras de decisión muestran cómo el clasificador divide el espacio de características en regiones correspondientes a cada clase.

```python
# Crear figura con dos subgráficos
_, axs = plt.subplots(ncols=2, figsize=(12, 5))

for ax, weights in zip(axs, ("uniform", "distance")):
    # Entrenar el modelo con el esquema de ponderación actual
    clf.set_params(knn__weights=weights).fit(X_train, y_train)
    
    # Visualizar las fronteras de decisión
    disp = DecisionBoundaryDisplay.from_estimator(
        clf,
        X_test,
        response_method="predict",
        plot_method="pcolormesh",
        xlabel=iris.feature_names[0],
        ylabel=iris.feature_names[1],
        shading="auto",
        alpha=0.5,
        ax=ax,
    )
    
    # Superponer los puntos de datos
    scatter = disp.ax_.scatter(
        X.iloc[:, 0], 
        X.iloc[:, 1], 
        c=y, 
        edgecolors="k"
    )
    
    # Añadir leyenda
    disp.ax_.legend(
        scatter.legend_elements()[0],
        iris.target_names,
        loc="lower left",
        title="Classes",
    )
    
    # Título del subgráfico
    _ = disp.ax_.set_title(
        f"3-Class classification\n(k={clf[-1].n_neighbors}, weights={weights!r})"
    )

plt.tight_layout()
plt.show()
```

---
### 8. Análisis de Resultados

#### 8.1 Interpretación de las Fronteras de Decisión

**Observaciones esperadas:**

- **Uniform weights**: Las fronteras tienden a ser más suaves y generalizadas
- **Distance weights**: Las fronteras pueden ser más irregulares y adaptarse mejor a puntos locales

#### 8.2 Métricas de Rendimiento

```python
# Evaluar ambos modelos con métricas detalladas
for weights in ("uniform", "distance"):
    clf.set_params(knn__weights=weights).fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    
    print(f"\n{'='*60}")
    print(f"Resultados con weights='{weights}'")
    print(f"{'='*60}")
    
    print("\nMatriz de Confusión:")
    print(confusion_matrix(y_test, y_pred))
    
    print("\nReporte de Clasificación:")
    print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

---
### 9. Conclusiones

**Puntos clave del análisis:**

1. **Preprocesamiento**: El escalado de características es crucial para KNN debido a su dependencia de distancias

2. **Selección de k**: Se utilizó k=11, un valor relativamente alto que ayuda a reducir el sobreajuste

3. **Ponderación**:
   - *Uniform*: Apropiado cuando todos los vecinos son igualmente confiables
   - *Distance*: Útil cuando los vecinos más cercanos deben tener mayor influencia

4. **Limitaciones**: Al usar solo dos características, perdemos información que podría mejorar la clasificación

**Recomendaciones para mejorar el modelo:**

- Experimentar con diferentes valores de k
- Incluir todas las características disponibles
- Probar métricas de distancia alternativas (Manhattan, Minkowski)
- Realizar validación cruzada para una evaluación más robusta
